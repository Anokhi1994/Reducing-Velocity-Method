```python
import math

tol = 1e-6

#Transformed_Components_Rotation[abs(Transformed_Components_WI_Rotation) < tol] = 0.0
#Transformed_Components_Spud_Rotation[abs(Transformed_Components_WI_Rotation) < tol] = 0.0

st5_curlx_SurfaceWave = (st5_curlx[0].data[300:500])/2  #Surface Wave Rotations  X Component
#st5_curlx_SurfaceWave[abs(st5_curlx_SurfaceWave) < tol] = np.nan
st5_times_SurfaceWave = st5_curlx[0].times()[300:500]    #Time Axis
st5_zv_SurfaceWave = -st5_zv[0].data[300:500]              #Surface Wave Velocity Z component
st5_zv_SurfaceWave[abs(st5_zv_SurfaceWave) < tol] = 1
st5_ratio_Rx_Vz = (st5_curlx_SurfaceWave/(np.cos(45/180*math.pi)))/-st5_zv_SurfaceWave # (Rx)/(Vz) 
                                                                                        #Slowness Vector

fig = plt.figure(figsize=(14,4))   #Plotting Rotations Surface Wave X component
fig.add_subplot()
plt.plot(st5_times_SurfaceWave, st5_curlx_SurfaceWave, color = 'b',label='Rotation_Surface_Wave')
plt.plot(st5_times_SurfaceWave, -st5_zv_SurfaceWave, color = 'g',label='Velocity_Surface_Wave')
plt.legend()
plt.title("X Component")

fig = plt.figure(figsize=(14,4))    #Plotting Velocity Surface Wave Z Component
fig.add_subplot()
plt.plot(st5_times_SurfaceWave, -st5_curlx_SurfaceWave, color = 'b',label='Rotation_Surface_Wave')
#plt.xlim([0.0, 1])
#plt.ylim([-100, 100])
plt.legend()
plt.title("Z Component")


fig = plt.figure(figsize=(14,4))   #Plotting Slowness vector 
fig.add_subplot()
plt.plot(st5_times_SurfaceWave, (st5_ratio_Rx_Vz*math.cos(45*180/math.pi)), color = 'b',label='Slowness')
plt.legend()
plt.title("Slowness")

fig = plt.figure(figsize=(14,4))   #Plotting Phase Velocity vector 
fig.add_subplot()
plt.plot(st5_times_SurfaceWave, 1/(st5_ratio_Rx_Vz), color = 'b',label='Phase Velocity')
plt.legend()
plt.title("Phase Velocity")
```
